```python
# Q.9) Replace null values in columns with suitable replacements


# to replace null values of last review column we are appplying the forward fill

# Comment:because its data is less than 30% missing we are applying forward fill to the column last review to fill the null values
df['last review'].ffill(inplace=True)
```

```python
# to replace null values of last review column we are appplying the forward fill

# COMMENT : since the column reviews_per_month is continous and is positively skewed with less than 30% missing values the median imputation is best

df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].median())
```

```python
# to relace null values of hostname and name column

# comment : we are applying the mode imputation in both columns because it has categorical data

df['name'] = df['name'].fillna(df['name'].mode()[0])
df['host_name'] = df['host_name'].fillna(df['host_name'].mode()[0])
```
